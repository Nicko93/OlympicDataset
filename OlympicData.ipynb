```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Create a DataFrame for the top 10 countries
top_10_countries = athletes_df['Team'].value_counts().sort_values(ascending=False).head(10).reset_index()
top_10_countries.columns = ['Country', 'Count']

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world map with the top 10 countries DataFrame
world = world.merge(top_10_countries, left_on='name', right_on='Country')

# Plot the world map
fig, ax = plt.subplots(figsize=(15, 10))
world.boundary.plot(ax=ax)
world.plot(column='Count', legend=True, ax=ax, cmap='OrRd', legend_kwds={'label': "Number of Athletes"})
plt.title('Top 10 Countries Participating in the Olympics')
plt.show()
```

```python
#Country Participating in Olympics
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Assuming athletes_df contains the necessary data

# Create a DataFrame for the country counts
country_counts = athletes_df['Team'].value_counts().reset_index()
country_counts.columns = ['Country', 'Count']

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world map with the country counts DataFrame
world = world.merge(country_counts, left_on='name', right_on='Country', how='left')

# Fill missing values with 0
world['Count'] = world['Count'].fillna(0)

# Plot the world map
fig, ax = plt.subplots(figsize=(15, 10))
world.boundary.plot(ax=ax)
world.plot(column='Count', legend=True, ax=ax, cmap='OrRd', legend_kwds={'label': "Number of Athletes"})
plt.title('Countries Participating in the Olympics')
plt.show()
```
